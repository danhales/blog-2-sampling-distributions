```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm

plt.style.use('seaborn-dark')
```

```python
def get_heights():
    return pd.read_csv('FEV.csv')['height']
```

```python
heights = get_heights()
heights
```

```python
plt.hist(heights)
plt.xlabel('Height (in)')
plt.ylabel('Frequency')
plt.title('Height of Study Participants')
plt.show()
```

```python
sample1 = heights.sample(5)
mean1 = sample1.mean()

plt.hist(sample1)
plt.xlabel('Height (in)')
plt.ylabel('Frequency')
plt.title('Height of Study Participants')
plt.show()

print(mean1)
```

```python
def plot_samples(sample_size, number_of_samples):
    for _ in tqdm.trange(number_of_samples):
        plt.hist(heights.sample(sample_size), alpha=0.1, color='blue')
    plt.xlabel('Height (in)')
    plt.ylabel('Frequency')
    plt.title('Height of Study Participants')
    plt.show()
```

```python
plot_samples(sample_size=600, number_of_samples=10)
```
