```python
import pandas as pd
import numpy as np
import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
figure_size = (10,5)

bar_color = '#4682B4'

plt.style.use('seaborn-muted')
```

Our exploration uses the dataset FEV.csv, which is also found in this repository. We'll look at only one column: height.

```python
# remember: this Series contains our entire population.
heights = pd.read_csv('FEV.csv')['height']
len(heights)
```

```python
# take a sample of size 5, and find the sample mean
# random_state=0 will allow you to get the same random sample as me
sample1 = heights.sample(n=5, random_state=1)
sample1.mean()
```

```python
# take another random sample.
sample2 = heights.sample(n=5, random_state=2)
sample2.mean()
```

```python
# take ten samples
ten_samples = [heights.sample(n=5, random_state=i) for i in range(1, 11)]
means = [sample.mean() for sample in ten_samples]
plt.figure(figsize=figure_size)
plt.grid(b=False)
plt.hist(means, bins=10, color='#4682B4')
plt.axvline(sample1.mean(), color='darkslategray', linestyle='--', linewidth=3)
plt.xlabel('Sample Mean (in.)', size=18)
plt.ylabel('Frequency', size=18)
plt.yticks([0,1,2,3])
plt.title(r'Distribution of $\bar{x}$ for n=5 across 10 samples', size=20)
plt.savefig('xbar-n-equals-5-10-samples.png')
plt.show()
```

```python
ten_thousand_samples = [heights.sample(n=5, random_state=i) for i in tqdm.trange(1, 10001)]
means = [sample.mean() for sample in ten_thousand_samples]
plt.figure(figsize=figure_size)
plt.grid(b=False)
plt.hist(means, bins=10, color='#4682B4')
plt.xlabel('Sample Mean (in.)', size=18)
plt.ylabel('Frequency', size=18)
plt.yticks(range(500,3001, 500))
plt.title(r'Distribution of $\bar{x}$ for n=5 across 10,000 samples', size=20)
plt.axvline(np.mean(means), color='blue', linestyle='--', linewidth=3)
plt.axvline(sample1.mean(), color='darkslategray', linestyle='--', linewidth=3)
plt.savefig('xbar-n-equals-5-10000-samples.png')
plt.show()
```

```python
plt.figure(figsize=figure_size)
plt.grid(b=False)
plt.hist([heights.sample(5, random_state=i).mean() for i in tqdm.trange(1, 10001)], 10, alpha=0.2, color='#4682B4')
plt.hist([heights.sample(50, random_state=i).mean() for i in tqdm.trange(1, 10001)], 10, alpha=0.5, color='#4682B4')
plt.hist([heights.sample(500, random_state=i).mean() for i in tqdm.trange(1, 10001)], 10, alpha=1, color='#4682B4')
plt.axvline(heights.mean(), 0, 1, color='blue', linestyle='--')
plt.legend(['n=5', 'n=50', 'n=500'], prop={'size':14})
plt.xlabel('Sample Mean (in.)', size=14)
plt.ylabel('Frequency', size=18)
plt.yticks(range(500,3001, 500))
plt.title(r'Distribution of $\bar{x}$ across 10,000 samples', size=20)
plt.savefig('three-sample-sizes-graph.png')
plt.show()
```
